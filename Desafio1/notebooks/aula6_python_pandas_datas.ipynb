#**Trabalhando com Planilhas do Excel**

```python
#importar a biblioteca pandas
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#Exibindo as 5 primeiras linhas
df2.head()
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as 5 primeiras linhas
df.head()
```

```python
#Exibindo as 5 ultimas linhas
df.tail()
```

```python
#AAmostra das 5 primeiras linhas
df.sample(5)
```

```python
#Mostraando o tipo de dados 
df.dtypes
```

```python
#Alterando o tipo de daddo da coluna lojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

#**Tratando valores faltantes**

```python
#Consulta linhas com valores faltantes
df.isnull().sum()
```

```python
#Substiruindo os valores nulos pela MEDIA
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#Valor da MEDIA
df["Vendas"].mean()
```

```python
#Substitui o valor de Vendas por zero(0)
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apaga as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apaga as linhas com valores nulos em uma coluna especifica
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

#**Criando Novas Colunas**

```python
#Criando coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"]) 
```

```python
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior Receita
df["Receita"].max()
```

```python
#Retornando a maior Receita
df["Receita"].min()
```

```python
#nlargest - top 3 da coluna REceita
df.nlargest(3, "Receita")
```

```python
#nsmallest - top 3 das piores Receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordena o conjunto de dados - 10 maiores Receitas
df.sort_values("Receita", ascending=False).head(10)
```

#**Trabalhando com Datas**

```python
#Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#Verifica o tipo de dado de cada coluna
df.dtypes
```

```python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento da Receita por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando uma nova coluna com o ano
df["Ano_Venda"] =  df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraindo o mês e o dia
df["mes_venda"],  df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Retornando a data mais antiga
df["Data"].min()
```

```python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando uma coluna de semestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
